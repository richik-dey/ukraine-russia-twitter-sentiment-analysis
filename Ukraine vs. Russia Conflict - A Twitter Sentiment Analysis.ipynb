### **2.4 Data Cleaning**

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def clean_tweet(text):
    text = str(text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'@[\w]*', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text.lower()

def is_english(text):
    words = set(text.split())
    return len(words & ENGLISH_STOP_WORDS) > 2

df['clean_text'] = df['text'].apply(clean_tweet)
df = df[df['clean_text'].apply(is_english)]
df[['text', 'clean_text']].head()
```

### **2.5 VADER Sentiment Analyzer**

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
scores = []
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

for tweet in df['clean_text']:
    score = analyzer.polarity_scores(tweet)
    pos = score['pos']
    neu = score['neu']
    neg = score['neg']
    compound = score['compound']
    scores.append({
        "Positive": pos,
        "Negative": neg,
        "Neutral": neu,
        "Compound": compound})
    compound_list.append(compound)
    positive_list.append(pos)
    negative_list.append(neg)
    neutral_list.append(neu)

sentiments_score = pd.DataFrame(scores)
df = pd.concat([df.reset_index(drop=True), sentiments_score], axis=1)
df.head()
```

### **2.6 Classify Tweets into Positive, Negative and Neutral Category**

```python
conditions = [
    df['Compound'] <= -0.5,
    (df['Compound'] > -0.5) & (df['Compound'] < 0.5),
    df['Compound'] >= 0.5]

values = ['Negative', 'Neutral', 'Positive']
df['Category'] = np.select(conditions, values, default='Neutral')
df[['clean_text', 'Category']].head()
```

# **7. Results and Analysis**

### **7.1 ROC Curve Comparison for All Models**

```python
from sklearn.preprocessing import label_binarize

classes = [0, 1, 2]
Y_test_bin = label_binarize(y_test, classes=classes)
n_classes = Y_test_bin.shape[1]

y_score_log = log_model.predict_proba(X_test_tfidf)
y_score_nb = nb_model.predict_proba(X_test_tfidf)
y_score_rf = rf_model.predict_proba(X_test_tfidf)
y_score_dt = dt_model.predict_proba(X_test_tfidf)
y_score_knn = knn_model.predict_proba(X_test_tfidf)
y_score_svc = svc_model.decision_function(X_test_tfidf)
```

```python
from sklearn.metrics import roc_curve, auc

model_scores = [
    ('Logistic Regression', y_score_log),
    ('Naive Bayes', y_score_nb),
    ('Random Forest', y_score_rf),
    ('Decision Tree', y_score_dt),
    ('KNN', y_score_knn),
    ('Linear SVC', y_score_svc)]

plt.figure(figsize=(8, 5))
colors = ['blue', 'green', 'purple', 'orange', 'brown', 'red']

for (label, score), color in zip(model_scores, colors):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(Y_test_bin[:, i], score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= n_classes
    macro_auc = auc(all_fpr, mean_tpr)
    plt.plot(all_fpr, mean_tpr, color=color, label=f'{label} (AUC = {macro_auc:.2f})', lw=2)

plt.plot([0,1], [0,1], 'k--', lw=2, label='Chance')
plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.show()
```

# **8. Deep Learning Algorithms**

### **8.1 CNN (Convolutional Neural Network)**

```python
import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

max_words = 10000
max_len = 50

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')

cnn_model = Sequential([
    Embedding(max_words, 128, input_length=max_len),
    Conv1D(128, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')])

cnn_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])

history_cnn = cnn_model.fit(
    X_train_pad, y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=128,
    verbose=1)

loss, acc = cnn_model.evaluate(X_test_pad, y_test, verbose=0)
print(f"CNN Test Accuracy: {acc*100:.2f}%")
```

### **8.2 LSTM (Long Short-Term Memory)**

```python
lstm_model = Sequential([
    Embedding(max_words, 128, input_length=max_len),
    LSTM(64, return_sequences=False),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')])

lstm_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])

history_lstm = lstm_model.fit(
    X_train_pad, y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=128,
    verbose=1)

loss, acc = lstm_model.evaluate(X_test_pad, y_test, verbose=0)
print(f"LSTM Test Accuracy: {acc*100:.2f}%")
```

### **8.3 GRU (Gated Recurrent Unit)**

```python
gru_model = Sequential([
    Embedding(max_words, 128, input_length=max_len),
    GRU(64, return_sequences=False),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')])

gru_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])

history_gru = gru_model.fit(
    X_train_pad, y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=128,
    verbose=1)

loss, acc = gru_model.evaluate(X_test_pad, y_test, verbose=0)
print(f"GRU Test Accuracy: {acc*100:.2f}%")
```

### **8.4 BiLSTM (Bidirectional LSTM)**

```python
bilstm_model = Sequential([
    Embedding(max_words, 128, input_length=max_len),
    Bidirectional(LSTM(64, return_sequences=False)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')])

bilstm_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])

history_bilstm = bilstm_model.fit(
    X_train_pad, y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=128,
    verbose=1)

loss, acc = bilstm_model.evaluate(X_test_pad, y_test, verbose=0)
print(f"BiLSTM Test Accuracy: {acc*100:.2f}%")
```

### **8.5 CNN + LSTM Hybrid**

```python
hybrid_model = Sequential([
    Embedding(max_words, 128, input_length=max_len),
    Conv1D(128, 5, activation='relu'),
    MaxPooling1D(pool_size=2),
    LSTM(64, return_sequences=False),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')])

hybrid_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])

history_hybrid = hybrid_model.fit(
    X_train_pad, y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=128,
    verbose=1)

loss, acc = hybrid_model.evaluate(X_test_pad, y_test, verbose=0)
print(f"CNN+LSTM Test Accuracy: {acc*100:.2f}%")
```

### **8.6 TCN (Temporal Convolutional Network)**

```python
from tensorflow.keras.layers import Conv1D, Dense, Dropout, Input, Add
from tensorflow.keras.models import Model

def tcn_block(inputs, filters, kernel_size=3, dilation_rate=1):
    conv = Conv1D(filters, kernel_size, dilation_rate=dilation_rate, padding='causal')(inputs)
    conv = Activation('relu')(conv)
    conv = Dropout(0.2)(conv)
    return conv

input_layer = Input(shape=(max_len,))

x = Embedding(max_words, 128)(input_layer)
x1 = tcn_block(x, 64, dilation_rate=1)
x2 = tcn_block(x, 64, dilation_rate=2)
x3 = tcn_block(x, 64, dilation_rate=4)

x = Add()([x1, x2, x3])
x = GlobalAveragePooling1D()(x)
x = Dropout(0.5)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.5)(x)
outputs = Dense(3, activation='softmax')(x)

tcn_model = Model(inputs=input_layer, outputs=outputs)

tcn_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])

history_tcn = tcn_model.fit(
    X_train_pad, y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=128,
    verbose=1)

loss, acc = tcn_model.evaluate(X_test_pad, y_test, verbose=0)
print(f"TCN Test Accuracy: {acc*100:.2f}%")
```

### **9.1 Accuracy and Loss Plots**

```python
def plot_acc_loss(history, model_name="Model"):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy', linewidth=2)
    plt.plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title(f'{model_name} - Accuracy Over Epochs')
    plt.grid(True, alpha=0.3)

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss', linewidth=2)
    plt.plot(history.history['val_loss'], label='Val Loss', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title(f'{model_name} - Loss Over Epochs')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

plot_acc_loss(history_cnn, "CNN")
plot_acc_loss(history_lstm, "LSTM")
plot_acc_loss(history_gru, "GRU")
plot_acc_loss(history_bilstm, "BiLSTM")
plot_acc_loss(history_hybrid, "CNN+LSTM")
plot_acc_loss(history_tcn, "TCN")
```

### **9.2 Predict Sentiment of New Tweet**

```python
def predict_new_tweet():

    analyzer = SentimentIntensityAnalyzer()
    while True:
        text = input('Enter a message: ').strip()
        if text.lower() == 'exit':
            print("Exiting prediction.")
            break
        if not text:
            print("Please enter a non-empty message.")
            continue
        s = analyzer.polarity_scores(text)
        if s['compound'] > 0.5:
            res = 'Positive'
        elif s['compound'] < -0.5:
            res = 'Negative'
        else:
            res = 'Neutral'
        print(f'Sentiment: {res}\nBreakdown: {s}\n')

predict_new_tweet()
```
